# Entanglement Suite (suite3)

The entanglement pillar controls **space–time coupling** in the triggering kernel.

| Level | Kernel | Wave speed |
|-------|--------|-----------|
| L0 | separable ExponentialGaussian | 0 (baseline) |
| L1 | traveling wave | 0.05 |
| L2 | traveling wave | 0.15 |
| L3 | traveling wave | 0.30 |

Increasing wave speed makes the triggering kernel non-separable: offspring
are displaced in the NE direction (θ = π/4) relative to their parent.


```python
from hawkesnest.suites import EntanglementSuite
from hawkesnest.viz import plot_kde_panels, plot_suite_event_grid, plot_events_3d
import matplotlib.pyplot as plt

suite = EntanglementSuite()
results = [suite.generate(level=lv, n_events=300, seed=0) for lv in suite.levels()]
{r.level: r.n_events for r in results}
```

## Primary — smooth KDE density across levels

At L0 the density tracks the three-cluster background. As wave speed increases (L1–L3), the KDE elongates in the NE direction, revealing the traveling-wave excitation pattern.

```python
fig = plot_kde_panels(results, cmap="magma",
                      title="Entanglement suite — KDE spatial density (L0 → L3)")
plt.show()
```

## Secondary — 2D event scatter coloured by arrival time

```python
fig = plot_suite_event_grid(results, color_by="t", cmap="viridis",
                            title="Entanglement suite — event scatter (colour = time)")
plt.show()
```

## Tertiary — 3D (x, y, t) cloud for L3

The strongest wave (L3) shows a visible NE directional streak when the time axis is added.

```python
fig, ax = plot_events_3d(results[-1].events, color_by="t",
                         title="Entanglement L3 — (x, y, t)")
plt.show()
```
