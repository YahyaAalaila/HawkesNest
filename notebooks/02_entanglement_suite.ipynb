# Entanglement Suite

Generate tiny examples for all stable suite3 levels using the public API.

```python
from hawkesnest.suites import EntanglementSuite

suite = EntanglementSuite()
results = {
    level: suite.generate(level=level, n_events=40, seed=0)
    for level in suite.levels()
}
{level: result.events.shape for level, result in results.items()}
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, len(results), figsize=(12, 3), sharex=True, sharey=True)
for ax, (level, result) in zip(axes, results.items()):
    ax.scatter(result.events["x"], result.events["y"], c=result.events["t"], s=18, cmap="viridis")
    ax.set_title(level)
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
plt.tight_layout()
plt.show()
```
